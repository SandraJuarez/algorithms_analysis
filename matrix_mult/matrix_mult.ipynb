```python
import time
import random
import numpy as np
import matplotlib.pyplot as plt
def measure_average_complexity(dimensions, num_tests):
    average_times = []

    for n in dimensions:
        times = []

        for _ in range(num_tests):
            # Generate random dimensions for matrices
            dimensions = [random.randint(1, 10) for _ in range(n + 1)]

            # Measure the execution time
            start_time = time.time()
            matrix_mult_chain_order(dimensions)
            end_time = time.time()

            times.append(end_time - start_time)

        average_time = np.mean(times)
        average_times.append(average_time)

    return average_times

# Specify the range of dimensions and number of tests
dimensions = list(range(2, 100))  # Dimensions from 2 to 10
num_tests = 10

average_times = measure_average_complexity(dimensions, num_tests)

# Plot the average time complexity
plt.plot(dimensions, average_times)
plt.xlabel('Number of Matrices')
plt.ylabel('Average Execution Time (s)')
plt.title('Average Complexity of Matrix Chain Multiplication')
plt.show()
```
